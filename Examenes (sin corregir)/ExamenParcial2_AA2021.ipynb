```python
def is_prime(n):
    """
    Donat un enter 'n' retorna True si és un nombre primer o False si no ho és
    """
    if n < 2: 
         return False;
    if n % 2 == 0:             
         return n == 2
    k = 3
    while k*k <= n:
        if n % k == 0:
            return False
        k += 2
    return True
```
